##### Handling Missing and Incorrect Values
 - Lmp
    - Missing Values will be replaced with 'Not Recorded' for females and 'Not Applicable' for males
    - Some rows where the client is a male have an LMP which is not applicable and these will be updated with 'Not Applicable'
    - Some Female Lmp records are unrealistic eg Lmp being 1900-01-01: These will be updated with 'Not Recorded'

```python
# Replacing LMP for males with "Not Applicable"
df.loc[df['Gender'].str.lower()=='m','Lmp']= 'Not Applicable'
```

```python
# Replacing Incorrect Female Lmp Records
df.loc[df['Lmp']=='1900-01-01','Lmp']= 'Not Recorded'
```

```python
# Replacing Null Female Lmp Records
df.loc[(df['Gender'].str.lower()=='f') & (df['Lmp'].isna()), 'Lmp']='Not Recorded'
```

##### Handling Missing Values
- BP
    - Some values were way out of range: 
        - Systolic levels that were humanly out of range: 50-300 
        - Diastolic levels out of human range: 20-200
    - Some BP records were incomplete
    - There were missing values
- The Cleaning Process will start by Spliting BP Records into BP Diastolic & BP Systolic

Note: Data Imputation for this column was questionable considering that Blood pressure is affected by so many factors whose data is not provided in the data set. 

```python
# Splitting BP into Diastolic and Systolic for Easier Cleaning Process
df[['Systolic','Diastolic']]=df['BP'].str.split('/',n=1,expand=True)
```

```python
#Replacing Systolic Missing Values (Do this before converting to numeric in order to use .str)
df.loc[df['Systolic'].str.strip()=="",'Systolic']=np.nan
```

```python
# Converting the split columns: Systolic into numeric format for ease of filtering out values out of range
df['Systolic']=pd.to_numeric(df['Systolic'], errors='coerce')
```

```python
# Replacing BP (Upper) Values that are out of range with NaN
df.loc[(df['Systolic']<50) | (df['Systolic']>300),'Systolic']= np.nan
```

```python
#Replacing Diastolic Missing Values (Do this before converting to numeric in order to use .str)
#df.loc[df['Diastolic'].str.strip()=="",'Diastolic']=np.nan
```

```python
#Converting Diastolic Column into numeric format (inconsistent values that have /// will be replaced with null)
df['Diastolic']=pd.to_numeric(df['Diastolic'],errors='coerce')
```

```python
#Replacing out of range BP Lower Values/Diastolic with nan
df.loc[(df['Diastolic']<20)|(df['Diastolic']>200),'Diastolic']=np.nan
```

```python
#Filling the Null Values using mode 
df['Systolic']=df['Systolic'].fillna(df['Systolic'].median())
df['Diastolic']=df['Diastolic'].fillna(df['Diastolic'].median())
```

```python
#Merging the two columns back together into BP_
df['BP_'] = df['Systolic'].astype(int).astype(str) + '/' + df['Diastolic'].astype(int).astype(str)
```

```python
#Droping the initial BP Column
df=df.drop('BP',axis=1)
```

#### Cleaning Height
    - Judging from the nature of the values provided in the height column;
        - values between 0.5 and 3 were assumed to be in m and were converted to cm
        - Values between 3.1 to 8 were assumed to be in Ft.
        - Values between 35 to 300 were assumed to be in cm and were kept as they are

```python
#Cleaning Height
def clean_height(h):
    if 0.5 <= h < 3:  
        return h * 100
    elif 3.1 <= h <= 8:   
        return h * 30.48
    elif 35 <= h <= 300:
        return h
    else:
        return np.nan

df['Height'] = df['Height'].apply(clean_height)
```

```python
#Drop the generated null height values
df=df.dropna()
```

#### Feature Engineering
    - BMI 
    - Weight Categories

```python
# BMI          Note that the height is in cm
df['BMI']=(df['Weight_']/(df['Height']/100)**2).round(1)
```

```python
# Categorizing BMI
def BMI_Cats(BMI):
    if BMI <18.5:
        return 'Underweight'
    elif 18.5<=BMI<=24.9:
        return 'Normal_weight'
    elif 25<=BMI<=29.9:
        return 'Overweight'
    elif BMI>=30:
        return 'Obese'
    else:
        return 'N/A'
    
df['BMI_Categories']=df['BMI'].apply(BMI_Cats)
```
